```python
#Task 1: Python Basics & Operators

# Step 1: Taking two numbers as input from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Step 2: Performing basic arithmetic operations
add = num1 + num2
sub = num1 - num2
mul = num1 * num2

# Checking to avoid division by zero
if num2 != 0:
    div = num1 / num2
    mod = num1 % num2
else:
    div = "Undefined (division by zero)"
    mod = "Undefined (modulus by zero)"

pow = num1 ** num2

# Step 3: Performing comparison operations
greater_than = num1 > num2
less_than = num1 < num2
equal_to = num1 == num2
not_equal_to = num1 != num2

# Step 4: Performing logical operations
logical_and = num1 and num2
logical_or = num1 or num2
logical_not_1 = not num1
logical_not_2 = not num2

# Step 5: Displaying the results using formatted output
print("\n----- Arithmetic Operations -----")
print(f"{num1} + {num2} = {add}")
print(f"{num1} - {num2} = {sub}")
print(f"{num1} * {num2} = {mul}")
print(f"{num1} / {num2} = {div}")
print(f"{num1} % {num2} = {mod}")
print(f"{num1} ** {num2} = {pow}")

print("\n----- Comparison Operations -----")
print(f"{num1} > {num2}  = {greater_than}")
print(f"{num1} < {num2}  = {less_than}")
print(f"{num1} == {num2} = {equal_to}")
print(f"{num1} != {num2} = {not_equal_to}")

print("\n----- Logical Operations -----")
print(f"({num1}) and ({num2}) = {logical_and}")
print(f"({num1}) or ({num2})  = {logical_or}")
print(f"not ({num1}) = {logical_not_1}")
print(f"not ({num2}) = {logical_not_2}")
```

```python
#Task 2: Lists and Arrays

# Importing required libraries
import random
import numpy as np

# Step 1: Creating a list of 10 random integers
numbers = random.sample(range(1, 101), 10)
print("Original list:", numbers)

# Step 2: Adding and removing elements
numbers.append(50)          # Add a number to the list
numbers.remove(numbers[0]) # Remove the first element from the list
print("After adding and removing:", numbers)

# Step 3: Finding maximum and minimum values
max_value = max(numbers)
min_value = min(numbers)
print("Maximum value:", max_value)
print("Minimum value:", min_value)

# Step 4: Sorting the list
numbers.sort()
print("Sorted list:", numbers)

# Step 5: Converting the list into a NumPy array
array = np.array(numbers)
print("NumPy array:", array)

# Step 6: Calculating Mean, Median, and Standard Deviation
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)

# Print the results
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#Task 3: Dictionaries and Sets 

# Step 1: Creating a dictionary
student = {
    "name": "Hari",
    "course": "GenAI",
    "marks": 85
}

# Step 2: Adding grade based on marks
if student["marks"] >= 80:
    student["grade"] = "A"
elif student["marks"] >= 60:
    student["grade"] = "B"
else:
    student["grade"] = "C"

# Step 3: Printing all keys and values
for key, value in student.items():
    print(key, ":", value)

# Step 4: Creating two sets of AI tools
tools_set1 = {"ChatGPT", "TensorFlow", "PyTorch"}
tools_set2 = {"ChatGPT", "Scikit-learn", "Keras"}
print("AI Tools Set 1:", tools_set1)
print("AI Tools Set 2:", tools_set2)

# Finding common tools
common_tools = tools_set1 & tools_set2
print("Common tools:", common_tools)
```

```python
# Task 4: File Handling

# Step 1: Creating and opening a file in write mode
file = open("ai_students.txt", "w")

# Step 2: Writing student details into the file
file.write("Hari,85,A\n")
file.write("Ravi,72,B\n")
file.write("Anu,90,A\n")
file.write("Kiran,65,C\n")
file.write("Meena,80,A\n")

# Closing the file after writing
file.close()

# Step 3: Opening the file in read mode
file = open("ai_students.txt", "r")

# Reading all lines from the file
lines = file.readlines()

print("Students who scored above 75:")

# Checking each studentâ€™s marks
for line in lines:
    name, marks, grade = line.strip().split(",")
    if int(marks) > 75:
        print(name, marks, grade)

# Closing the file
file.close()
```
